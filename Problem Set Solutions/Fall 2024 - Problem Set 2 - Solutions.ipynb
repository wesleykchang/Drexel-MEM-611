2. Obtain the steady-state temperature distribution T(x) in a plane wall of thickness 2L and constant thermal conductivity k. Internal energy is generated at a rate of $\dot q(x) = \dot q_0[1-(x/L)^2] $

    The plane wall surfaces at x = L and x = -L are at a constant $T_w$


since it's a plane wall and no other dimensions provided, we assume it is a 1D, steady-state problem. 

use the general equation in cartesian coordinates: $\frac{d^2T}{dx^2}\ +\ \frac{\dot{q}}{k}(1-(x/L)^2)\ =\ 0$

with boundary conditions: $T(L)\ =\ T_w$, $dT(0)/dx\ =\ 0$

due to symmetry, we know there is no change in temperature gradient in the middle of the cylinder

$T(x)\ =\ -\dot{q}/k (x^2/2\ -\ x^4/12L^2)\ +\ c_1 x\ +\ c_2$

$c_1\ =\ 0$ and $c_2\ =\ T_w\ +\ \dot{q}/k * 5L^2/12$

$T(x)\ =\ T_w\ +\ \dot{q}L^2/2k (5/6\ +\ x^2/L^2\ -\ x^4/6L^4)$

3. Now we have a solid cylindrical rod with radius $r_0$. 
Derive the steady-state temperature distribution T(r), assuming constant thermal conductivity k and internal energy generation at a rate of $\dot{q}\ =\ \dot{q_0}[1-r/r_0]$
the surface of the cylinder is maintained at constant temperature $T_w$

$\frac{d^2 T}{dr^2}\ +\ \frac{1}{r}\frac{dT}{dr}\ +\ \frac{\dot{q}_0}{k}(1-r/r_0)\ =\ 0$

boundary conditions: $T(r_0)\ =\ T_w$, $T(0)\ =\ $ a defined value. 

move the internal heat generation term to the right side, multiply both sides by r to simplify LHS, and integrate both sides 

$r\frac{dT}{dr}\ =\ \int (r^2/r_0 -r)\dot{q}_0/k$

$T(r)\ =\ -\frac{\dot{q}_0}{k}(r^2/4\ -\ r^3/9r_0)\ +\ c_1ln(r)\ +\ c_2$

plug in B.C's: $c_1\ =\ 0$, $c_2\ =\ T_w\ +\ 5/36*\dot{q}_0 r_0^2/k$

therefore: $T(r) - T_w\ =\ \dot{q}r_0^2/k (5/36 - r^2/r_0^2 (1/4 - 1/9 * r/r_0))$

4. Derive an expression for the steady-state temperature distribution T(r) in a long solid cylinder with thermal conductivity k and radius $r_0$, and with internal energy generated at a rate of $\dot{q}$ per unit volume. The outer surface at r = $r_0$ is exposed to fluid at temperature $T_\infty$ with constant heat transfer coefficient h. 

- use the general equation in cylindrical coordinates

$\frac{d^2T}{dr^2}\ +\ \frac{1}{r}\frac{dT}{dr}\ + \frac{\dot{q}}{k}\ =\ 0$

boundary conditions:  $T(0)\ =\ $ a defined value, $-k\frac{dT(r_0)}{dr}\ =\ h(T(r_0)\ -\ T_\infty)$

solution: $T(r)\ =\ -\dot{q}r^2/4k\ +\ c_1ln(r)\ +\ c_2$

plug in B.C.'s: $c_1\ =\ 0$ in order for first BC to remain finite, cannot have log(0)

$c_2\ =\ T_\infty\ +\ \frac{\dot{q}r_0}{2h}\ +\ \frac{\dot{q}r_0^2}{4k}$

therefore:

$T(r)\ =\ T_\infty\ +\ \frac{\dot{q}r_0^2}{4k}(1-(r/r_0)^2)\ +\ \frac{\dot{q}r_0}{2h}$
