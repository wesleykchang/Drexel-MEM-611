
2. $\frac{d^2y}{dx^2}\ -\ \lambda ^2 y\ =\ 0$

$y(0)\ =\ 0,\ y(L)\ =\ 0$

first, find the solution by guessing:

guess $y\ =\ e^{\alpha x}\ =\ 0$

solve $\alpha^2 e^{\alpha x}\ -\ \lambda^2 e^{\alpha x}\ =\ 0$

$\alpha^2\ -\ \lambda^2\ =\ 0$

$\alpha\ =\ +/- \lambda$

therefore $y\ =\ c_1 e^{\lambda x}\ +\ c_2 e^{-\lambda x}$ i.e. y is a linear combination of the two solutions

given the definitions of hyberbolic sine and cosine, this could also be written as $y\ =\ c_3 sinh(\lambda x)\ +\ c_4 cosh(\lambda x)$ 

($sinh(\lambda x)\ =\ (e^{\lambda x}\ -\ e^{-\lambda x})/2$, $cosh(\lambda x)\ =\ (e^{\lambda x}\ +\ e^{-\lambda x})/2$)

remember that $sinh(0)\ =\ 0$ and $cosh(0)\ =\ 1$

plugging in boundary conditions:

$c_1\ +\ c_2\ =\ 0$ so $c_1\ =\ -c_2$

$c_1e^{\lambda L}\ +\ c_2e^{-\lambda L}\ =\ 0$

$c_2[e^{-\lambda L}\ -\ e^{-\lambda L}]\ =\ 0$

$c_2\ =\ 0$ so $c_1\ =\ 0$

therefore, only the trivial solution exists

similarly, if we used the hyberbolic sine and cosine form: 

$c_4 cosh(0)\ =\ 0$

$c_4\ =\ 0$

$c_3 sinh(\lambda L)\ =\ 0$

$\lambda\ =\ 0$ 

similar conclusion, only the trivial solution exists


3. 

$\frac{d^2 y}{dx ^2}\ +\ \lambda ^2 y\ =\ 0$

$dy(0)/dx \ =\ 0,\ dy(L)/dx\ =\ 0$

as in the example done in class, the solution is a linear combination of sine and cosine:

$y(x)\ =\ Asin(\lambda x)\ +\ Bcos(\lambda x)$

plugging in boundary conditions:

$\frac{dy}{dx}\ =\ A\lambda cos(\lambda x)\ -\ B\lambda sin(\lambda x)$

$\frac{dy(0)}{dx}\ =\ 0$ => $A\ =\ 0$

$\frac{dy(L)}{dx}\ =\ 0$ => $B\lambda sin(\lambda L)\ =\ 0$ 

$sin(\lambda L)\ =\ 0$ => $\lambda _n\ =\ n\pi/L,\ n\ =\ 0,1,2,...$, B is finite

the characteristic function is $\phi _n(x)\ =\ cos(\lambda _n x)$, $\lambda _n\ =\ n\pi/L, n\ =\ 0,1,2,...$

$T(0,y)\ =\ 0$

$T(a,y)\ =\ f(y)$

$T(x,0)\ =\ 0$

$T(x,b)\ =\ 0$

apply separation of variables and assume: $T(x,y)\ =\ X(x)*Y(y)$

note that the course example had the inhomogeneous boundary condition at T(x,b) rather than T(a,y)

$Y*\frac{\partial ^2 X}{\partial x^2}\ +\ \frac{1}{Y}\frac{\partial ^2 Y}{\partial y}\ =\ 0$

$-\frac{1}{X}\frac{\partial ^2X}{\partial x^2}\ =\ \frac{\partial ^2 Y}{\partial y} \frac{1}{Y}\ =\ -\lambda ^2$

$\frac{d^2 Y}{dy^2}\ +\ \lambda ^2 Y\ =\ 0$

$Y(y)\ =\ Asin(\lambda y)\ +\ Bcos(\lambda y)$ plugging in boundary conditions $Y(0)\ =\ 0,\ Y(b)\ =\ 0$

$Y(y)\ =\ A_n sin(\lambda _n y),\ \lambda _n\ =\ n\pi/b,\ n\ =\ 1,2,3,...$

$\frac{d^2X}{dx^2}\ -\ \lambda ^2 X\ =\ 0$

$X(x)\ =\ B_nsinh(\lambda _n x)$

$T(x,y)\ =\ \sum _{n=1}^\infty a_n sinh(\lambda _n x) sin(\lambda _n y)$
