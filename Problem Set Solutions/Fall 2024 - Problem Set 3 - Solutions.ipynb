1. This question asked to solve the temperature profile for the 2D x-y problem for the other two boundary conditions, i.e. we solved for the two cases where T = T0 at the top and bottom, now we need to find the temperature profile for the two cases where T = T0 at the left and right walls.


First let's solve for the case where the right wall at x = a is at a non-zero temperature $T_0$

a) Apply separation of variables and plug into the 2D steady-state X-Y general equation: 

$Y\frac{\partial ^2 X}{\partial x^2}\ +\ X\frac{\partial ^2 Y}{\partial y^2}\ =\ 0$

b) Separate and set equal to a constant: 

$-\frac{1}{X}\frac{\partial ^2 X}{\partial x^2}\ =\ \frac{\partial ^2 Y}{\partial y^2}\frac{1}{Y}\ =\ -\lambda ^2$

$\frac{\partial ^2 X}{\partial x^2}\ =\ \lambda^2 X$

$\frac{\partial ^2 Y}{\partial y^2}\ =\ -\lambda^2 Y$

c) Solve each differential equation as in the class example:

$X(x)\ =\ c_3 e^{\lambda x}\ +\ c_4 e^{-\lambda x}$

$Y(y)\ =\ c_1 sin(\lambda y)\ +\ c_2 cos(\lambda y)$

d) Plug in the boundary conditions to find the constants:

for $T(y=0)\ =\ 0$: $c_2\ =\ 0$, we divide by X and then plug in y = 0 which gets rid of the sine term in Y

for $T(y=b)\ = 0$: $c_1 sin(\lambda b)\ =\ 0$ which is periodically true when $\lambda\ =\ n\pi/b,\ n\ =\ 1,2,3,...$ recall that second term in Y was found to be zero in the prior boundary condition

for $T(x=0)\ = 0$: $c_3\ =\ -c_4$, we divide by Y and set X = 0

this leaves us with: $T\ =\ X*Y\ =\ c_4[e^{-\lambda x}\ -\ e^{\lambda x}]c_1 sin(\lambda y)\ =\ c_5 sin(\lambda y)sinh(\lambda x)$

note that this can be written in terms of a Fourier sine expansion: $\sum_{n=1}^{\infty}\ a_n sin(\lambda _n y)$ where $a_n\ =\ c_5 sinh(\lambda x)$

plug in the last boundary condition:

for $T(x=a)\ =\ T_0$: 

$T_0\ =\ \sum_{n=1}^\infty a_n sin(\lambda _n y)$, $a_n\ =\ c_5 sinh(\lambda a)\ =\ \frac{2}{b}\frac{b T_0}{n\pi} [1-(-1)^n]$

$a_n\ =\ \frac{2}{b} \int _0^b T_0 sin(n\pi y/b) dy$ by consulting Table 4.1 in Yener and Kakac for the normalization factor in the expansion coefficient and noting that we previously found what $\lambda$ is in the other boundary condition

after finding the expansion coefficient, plug back to solve for the constant $c_5$ at x = a:

$c_5\ =\ \frac{a_n}{sinh(\lambda a)}$ 

e) Plug in all solved constants to find the final temperature profile:

thus, $T\ =\ \frac{2T_0}{\pi} \sum _{n=1}^\infty \frac{1-(-1)^n}{n} \frac{sin(n\pi y/b) sinh(n\pi x/b)}{sinh(n\pi a/b)}$

f) Find the temperature at the desired point:

x/a = 0.25

y/b = 0.75

b = 2a

plug in, and should get $T/T_0\ =\ 0.165$


Now let's solve for the case where $T\ =\ T_0$ at the left wall.

a) Note that separation of variables gets us the exact same form as above:

$T\ =\ X*Y\ =\ [c_3 e^{\lambda x}\ +\ c_4 e^{-\lambda x}][c_1 sinh(\lambda y)\ +\ c_2cos(\lambda y)]$

b) The two boundary conditions at the top and bottom walls result in the same conclusions:

$T(y=0)\ =\ 0$: $c_2\ =\ 0$

$T(y=b)\ =\ 0$: $c_1sin(\lambda b)\ =\ 0$ where $\lambda\ =\ n\pi/b,\ n\ =\ 1,2,3,...$

c) At $T(x=a)\ =\ 0$: $c_3e^{\lambda a}\ +\ c_4e^{-\lambda a}\ =\ 0$ and rewriting in terms of hyberbolic sine and cosine: $c6\ =\ -c_5 \frac{sinh(\lambda a)}{cosh(\lambda a)}$ (since terms are rearranged, using different constant names just to be consistent)

Therefore, $T\ =\ c_1sin(\lambda y)c_5[sinh(\lambda x)\ -\ \frac{sinh(\lambda a)}{cosh(\lambda a)}cosh(\lambda x)]\ =\ c_7 sinh(\lambda y)(sinh(\lambda (a-x)))$

d) Using the last boundary condition:

$T(x=0)\ =\ T_0$: $T\ =\ \sum _{n=1}^\infty c_7 sin(\lambda y) sinh(\lambda (a-x))\ =\ \sum _{n=1}^{\infty} a_n sin(\lambda y)$, i.e. write in terms of a Fourier sine expansion

$a_n\ =\ c_7 sinh(\lambda (a-x))\ =\ \frac{2}{b} \int _0^b T_0 sin(n\pi y/b)dy$ solve for the expansion coefficient using the normalization factor in Table 4.1 and then solve for $c_7$ at x = 0: $c_7\ =\ a_n/sinh(n\pi a/b)$

Therefore, $T\ =\ \frac{2T_0}{\pi} \sum _{n=1}^{\infty} \frac{1-(-1)^n}{n} \frac{sin(n\pi y/b) sinh(n\pi(a-x)/b)}{sinh(n\pi a/b)}$

e) Find the temperature at the given point

x/a = 0.25

y/b = 0.75

b = 2a

$T/T_0\ =\ 0.64$









2. $\frac{d^2y}{dx^2}\ -\ \lambda ^2 y\ =\ 0$

$y(0)\ =\ 0,\ y(L)\ =\ 0$

first, find the solution by guessing:

guess $y\ =\ e^{\alpha x}\ =\ 0$

solve $\alpha^2 e^{\alpha x}\ -\ \lambda^2 e^{\alpha x}\ =\ 0$

$\alpha^2\ -\ \lambda^2\ =\ 0$

$\alpha\ =\ +/- \lambda$

therefore $y\ =\ c_1 e^{\lambda x}\ +\ c_2 e^{-\lambda x}$ i.e. y is a linear combination of the two solutions

given the definitions of hyberbolic sine and cosine, this could also be written as $y\ =\ c_3 sinh(\lambda x)\ +\ c_4 cosh(\lambda x)$ 

($sinh(\lambda x)\ =\ (e^{\lambda x}\ -\ e^{-\lambda x})/2$, $cosh(\lambda x)\ =\ (e^{\lambda x}\ +\ e^{-\lambda x})/2$)

remember that $sinh(0)\ =\ 0$ and $cosh(0)\ =\ 1$

plugging in boundary conditions:

$c_1\ +\ c_2\ =\ 0$ so $c_1\ =\ -c_2$

$c_1e^{\lambda L}\ +\ c_2e^{-\lambda L}\ =\ 0$

$c_2[e^{-\lambda L}\ -\ e^{-\lambda L}]\ =\ 0$

$c_2\ =\ 0$ so $c_1\ =\ 0$

therefore, only the trivial solution exists

similarly, if we used the hyberbolic sine and cosine form: 

$c_4 cosh(0)\ =\ 0$

$c_4\ =\ 0$

$c_3 sinh(\lambda L)\ =\ 0$

$\lambda\ =\ 0$ 

similar conclusion, only the trivial solution exists

$y(x)\ =\ Asin(\lambda x)\ +\ Bcos(\lambda x)$

$sin(\lambda L)\ =\ 0$ => $\lambda _n\ =\ n\pi/L,\ n\ =\ 0,1,2,...$, B is finite

the characteristic function is $\phi _n(x)\ =\ cos(\lambda _n x)$, $\lambda _n\ =\ n\pi/L, n\ =\ 0,1,2,...$
