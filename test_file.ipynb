```python
# Heating a Steel Rod

from matplotlib.animation import FuncAnimation


length = 10 #cm
k = 0.466
temp_L = 200
temp_R = 200
time = 10

dx = 0.1
dt = 0.01
x = np.linspace(0, length, int(length/dx))
t = np.linspace(0, time, int(time/dt))

u = np.zeros([len(t), len(x)])

u[0] = np.exp(-0.5 * ((x - length)/2)**2)
u[:, 0] = 200 
u[:, -1] = 200

figure1 = plt.figure()
ax1 = figure1.add_subplot(111)
ax1.plot(u[0,:])

def animate(t):
    ax1.clear()
    for x1 in range(1, len(x) - 1):
        u[t + 1, x1] = k * (dt/dx**2) *(u[t, x1 + 1] - 2*u[t, x1] + u[t, x1 - 1])

    ax1.plot(x1, t)
    ax1.set_xlabel('Distance along Rod')
    ax1.set_ylabel('Temperature')

ani = FuncAnimation(figure1, animate, frames = len(t) - 1, repeat = False)
f = 'animate_test_file.mp4'
ani.save(f)

plt.show()
```
