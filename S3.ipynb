## Back to thermo

Consider again a thin slice within a material with two different surface temperatures

First Law of Thermo: $$dE\ =\ \delta Q\ -\ \delta W $$

As $$\Delta t\ \Rightarrow 0: \frac{\delta E}{\delta t} = q - \dot{w}$$

Assumptions:

$$de\ =\ du $$
 all energy is internal energy, as there is no other source of energy (from chemical reactions, electrical input, etc)

$$du\ =\ \frac{\delta u}{\delta V}_T dV\ +\ c_v dT $$ 
for stationary solids (or incompressible fluids) of uniform composition

## Inserting these assumptions into the 1st Law

$$ \frac{\delta E}{\delta t} = q - \dot{w}$$ 

q = heat out - heat in, no work; $$q\ =\ \int _V \dot{q} * dV\ -\ \int _S\ -k\nabla T*ndS$$ 

LHS: $$ \frac{\delta E}{\delta t} = \frac{\delta}{\delta t} \int _V (\rho *c*dT) dV $$

RHS: $$q\ -\ \dot{w} =\ \int _V \dot{q} * dV\ -\ \int _S\ -k\nabla T*ndS $$

$$\int _V \rho c \frac{\delta T}{\delta t} dV\ -\ \int \dot{q} dV\ =\ \int _S\ -k\nabla T*ndS $$

## Divergence Theorem

$$ \int _S\ -k\nabla T*ndS =\ -\int _V \nabla k \nabla T dV$$


## General heat conduction equation (or Fourier Biot equation)

$$\int _V [-\rho c \frac{\delta T}{\delta t}\ +\ \nabla*k\nabla T\ + \dot{q}] dV\ =\ 0 $$

$$\rho c \frac{\delta T}{\delta t}\ =\ \nabla*k\nabla T\ +\ \dot{q} $$

$$\nabla^2 T + \frac{\dot{q}}{k} = \frac{1}{\alpha} \frac{\delta T}{\delta t}$$

thermal diffusivity: $$\alpha =\ \frac{k}{\rho c} $$ 


## Laplace Equation

$$\nabla^2 T = 0$$

## Full general equation in cartesian coordinates

$$\frac{\delta^2 T}{\delta x^2}\ +\ \frac{\delta^2 T}{\delta y^2}\ +\ \frac{\delta^2 T}{\delta z^2}\ =\ \frac{1}{\alpha}\ +\ \frac{\delta T}{\delta t} $$

## Cylindrical coordinates

$$\nabla^2 T\ =\ \frac{\delta^2T}{\delta r^2}\ +\ (1/r) \frac{\delta T}{\delta r}\ +\ (1/r^2) \frac{\delta^2T}{\delta \phi^2}\ +\ \frac{\delta^2T}{\delta z^2} $$

$$T\ =\ T(r,\phi,z,t) $$

$$x\ =\ rcos \phi,\ y\ =\ rsin \phi,\ z\ = z $$

How did we get this conversion?

First, write r and theta in terms of x and y.

$$x^2\ +\ y^2\ =\ r^2 $$

$$y/x\ =\ rsin(\theta)/rcos(\theta)\ =\ tan(\theta);\ \theta \ =\ tan^{-1}(y/x) $$

Next, split up the derivatives: $$dT/dx\ =\ \frac{dT}{dr}\frac{dr}{dx}\ +\ \frac{dT}{d\theta}\frac{d\theta}{dx} = cos(\theta)\frac{dT}{dr}\ -\ \frac{sin(\theta)}{r} \frac{dT}{dr}$$

Remember, we have to write r in terms of x and y to solve dr/dx:

$$dr/dx\ =\ 1/2(x^2\ +\ y^2)^{-1/2}*2x\ =\  1/2(r^2)^{-1/2}*2*rcos(\theta)\ =\ cos(\theta)$$

For the second derivatives:

$$d^2T/dx^2\ =\ d/dx(dT/dx)\ =\ d/dr(dT/dx)(dr/dx)\ +\ d/d\theta\ (dT/dx)(d\theta/dx) $$

Same thing for $$d^T/dy^2\ =\ (d/dy)(dT/dy)\ =\ (d/dr)(dT/dy)(dr/dy)\ +\ (d/d\theta)(dT/dy)(d\theta/dy)$$

Remember nothing changes for the z-axis

## One dimensional, steady-state

- One dimensional: heat transfer occurs only along a single coordinate axis 
- Steady-state: temperature at each point does not vary with time 

Remember, steady-state conditions with no internal energy source or sink means using the Laplace equation: 

$$\frac{d^2 T}{dx^2} =\ 0 $$

Solving by integration:

$$T(x)\ =\ C_1 x\ +\ C_2 $$

Use the following boundary conditions:

$$T(x=0)\ =\ T_1,\ T(x=L)\ =\ T_2$$

$$C_2\ =\ T_1 $$

$$C_1\ =\ \frac{T_2 - T_1}{L} $$

Therefore:

$$T(x)\ =\ (T_2\ -\ T_1)*x/L\ +\ T_1 $$

With constant thermal conductivity, the temperature distribution in the wall is linear with x. 

Rate of heat transfer calculated from Fourier's law of heat conduction: 

$$q\ =\ -kA(dT/dx)\ =\ kA(T_1 - T_2)/L $$

more complicated if k(T) or k(x) 

## Temperature dependent k

Recall now that most materials have thermal conductivity dependent on temperature...

From figure in Session 1, the temperature-dependent relationship is sort of linear...

$$k(T)\ =\  k_0*(1+\beta*T)$$

$$\int k(T)dT\ =\ -q/A*(x-0) $$



```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

x = np.linspace(0,1,100)
betas = [1, 0.7, -1, -5]

figure1 = plt.figure()
ax1 = figure1.add_subplot(111)

for beta in betas:
    T = 1-(1/beta)*(1-np.sqrt(1-beta*(2-beta)*x))
    ax1.plot(x,T)

plt.show()
```

## Hollow Cylinder

Now consider a hollow cylinder of inside radius r1 and outside radius r2. If cylinder is long enough so that end effects are neglected, then temperature distribution is one dimensional (as a function of r). 

$$d^2 T/dr^2\ =\ (1/r)(dT/dr) $$

$$T(r)\ =\ c_1 ln(r)\ +\ c_2 $$

boundary conditions:

$$T(r_1)\ =\ T_1 $$
$$T(r_2)\ =\ T_2 $$

substitute

$$T(r)\ =\ T_1 - (T_1 - T_2)/(ln(r_2/r_1)) ln(r/r_1) $$



# Example: Cone 

Consider a cone that is insulated except for the sides. The sides are at position x1 = 0.05m and x2 = 0.25 m with temperatures T1 = 400 K and T2 = 600 K. material thermal conductivity is k = 3.46 W/m-K. D (diameter of cone) = ax, where a = 0.25. 

(1) Schematic
 
draw it out

(2) Assumptions 

- steady-state
- 1D conduction
- no internal heat gen

(3) $$q\ =\ -kA*dT/dx\ =\ -k(\pi *D^2 / 4) * dT/dx$$


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

T1 = 400
T2 = 600
x1 = 0.05
x2 = 0.25 

x = np.linspace(0.05, 0.25, 1000)
T = T1 + (T1 - T2)*((1/x - 1/x1)/(1/x1 - 1/x2))

a = 0.25
D = a*x
area = np.pi*(D/2)**2

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(x[:-1], np.diff(T)/np.diff(x))
ax2.plot(x, 1/area, color='#E0001B')
plt.show()
```

## Example: Plane Wall

Consider a plane wall of thickness 2L and internal energy generated at a uniform rate ($$\dot{q} $$), and it is exposed to a fluid at temperature Tf on either side with constant heat transfer coefficient h. Heat flow is one dimensional, and k is constant. 

Recall from earlier the general heat conduction equation: $$\nabla^2 T + \frac{\dot{q}}{k} = \frac{1}{\alpha} \frac{\delta T}{\delta t}$$ 

1. Schematic

draw out

2. Assumptions

- steady-state conditions: $$dT/dt\ =\ 0 $$
- And in one dimension (x-axis):  $$d^2T/dx^2\ +\ \dot{q}/k\ =\ 0 $$

3. Initial and boundary equations

$$dT/dt\ =\ 0 $$

$$(dT/dx)_{x=0}\ =\ 0  $$

$$(k*\frac{dT}{dx} +\ hT)_{x=L}\ =\ hT_f $$

4. Solve

Integration: $$T(x)\ =\ -\frac{\dot{q}x^2}{2k}\ +\ C_1x\ +\ C_2 $$

Apply BCs: $$C_1\ =\ 0 $$

$$C_2\ =\ \frac{\dot{q}L^2}{2k}\ +\ \frac{\dot{q}L}{h}\ +\ T_f $$

=> $$T(x)\ =\ -\frac{\dot{q}x^2}{2k}\ +\ T_f\ +\ \frac{\dot{q}L}{h}\ +\ \frac{\dot{q}L^2}{2k} $$

Can be re-written as: $$\frac{T-T_f}{(\dot{q}L^2/2k)}\ =\ 1\ -\ (x/L)^2\ +\ 2/Bi  $$
where Bi = hL/k (Biot number)

The biot number is a dimensionless number that indicates the ratio of the thermal resistance for conduction inside a body to the resistance for convection at the surface. In other words, small Bi values indicates nearly uniform temperature fields inside (i.e. very high conductivity). Large Bi numbers indicate such temperature gradients are non-negligible
