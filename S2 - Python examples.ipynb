```python
## Heating a steel rod

## Initialize our constants
length = 10 # length of the rod
k = 0.466 # thermal conductivity 
temp_left = 200
temp_right = 400
total_time = 1000

dx = 0.1 
dt = 0.01

x_vec = np.linspace(0, length, int(length/dx))
t_vec = np.linspace(0, total_time, int(total_time/dt))

T = np.zeros([len(t_vec), len(x_vec)])

## Boundary conditions
T[:, 0] = temp_left
T[:, -1] = temp_right

for t in range(1, len(t_vec) - 1):
    for x in range(1, len(x_vec) - 1):
        T[t + 1, x] = T[t, x] + k * dt / dx**2 * (T[t, x+1] - 2*T[t,x] + T[t, x-1])

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(T[-1, :])
plt.show()

```
