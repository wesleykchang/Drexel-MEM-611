$\theta\ =\ T\ -\ T_i$

so $\frac{\partial ^2 \theta}{\partial x^2}\ =\ \frac{1}{\alpha} \frac{\partial \theta}{\partial t}$

$\theta (x,0)\ =\ 0$

$-k \frac{\partial \theta (x,0)}{\partial x}\ =\ q_w"$

$\frac{d^2 \overline{\theta}}{dx^2}\ -\ \frac{p}{\alpha} \overline{\theta}(x,p)\ =\ 0$

$c_1\ =\ 0$
